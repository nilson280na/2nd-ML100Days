### Hints: [隨機產生數值](https://blog.csdn.net/christianashannon/article/details/78867204)

```python
import pandas as pd
import numpy as np
```

```python
#從txt檔案讀取國家名稱

國家 = []
with open("data/country_list.txt", 'r') as f:
    data_tmp = f.readlines()

#去掉每一個element的'/n'

for i in data_tmp:
    國家.append(i.strip())

#隨機產生相對應的國家人口數

total_country = len(國家)
人口 = np.random.randint(10000, 1000000000, total_country)

#將國家和人口組成一個array

data = {'國家':國家, '人口':人口}

#格式化成一個data frame
final_data = pd.DataFrame(data)

pd.set_option('display.max_rows', total_country)
print(final_data)
```

```python
#找出人口數最多的國家

max_people_idx_num = final_data['人口'].idxmax()
data_content = final_data.iloc[max_people_idx_num]
print(data_content)
```
